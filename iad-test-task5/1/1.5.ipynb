```python
import numpy as np
from scipy import special
```

```python
sy, sx = face.shape
y, x = np.ogrid[0:sy, 0:sx]

centerx, centery = (660, 300) # центр изображения
a = 330 # полуось эллипса по оси x
b = 150 # полуось эллипса по оси y

mask = ((x - centerx) ** 2 / a ** 2 + (y - centery) ** 2 / b ** 2) > 1 # уравнение эллипса 
```

```python
face[mask] = 0
plt.imshow(face)
```
